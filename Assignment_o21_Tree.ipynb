```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def inOrderTraversal(node, values):
    if node is None:
        return

    inOrderTraversal(node.left, values)
    values.append(node.value)
    inOrderTraversal(node.right, values)

def convertToBST(node, values):
    if node is None:
        return

    convertToBST(node.left, values)
    node.value = values.pop(0)
    convertToBST(node.right, values)

def binaryTreeToBST(root):
    values = []
    inOrderTraversal(root, values)
    values.sort()
    convertToBST(root, values)




tree = TreeNode(10)
tree.left = TreeNode(2)
tree.right = TreeNode(7)
tree.left.left = TreeNode(8)
tree.left.right = TreeNode(4)


binaryTreeToBST(tree)


def printBST(root):
    if root is None:
        return

    printBST(root.left)
    print(root.value, end=' ')
    printBST(root.right)

printBST(tree)
```

```python
class newNode:

    def __init__(self, data):
        self.key = data
        self.left = None
        self.right = None

def insert(root, key):
    if root == None:
        root = newNode(key)
    elif root.key > key:
        root.left = insert(root.left, key)
    elif root.key < key:
        root.right = insert(root.right, key)
    return root

def distanceFromRoot(root, x):
    if root.key == x:
        return 0
    elif root.key > x:
        return 1 + distanceFromRoot(root.left, x)
    return 1 + distanceFromRoot(root.right, x)

def distanceBetween2(root, a, b):
    if root == None:
        return 0

    if root.key > a and root.key > b:
        return distanceBetween2(root.left, a, b)

    if root.key < a and root.key < b:
        return distanceBetween2(root.right, a, b)

    if root.key >= a and root.key <= b:
        return (distanceFromRoot(root, a) +
                distanceFromRoot(root, b))

def findDistWrapper(root, a, b):
    if a > b:
        a, b = b, a
    return distanceBetween2(root, a, b)


root = None
root = insert(root, 20)
insert(root, 10)
insert(root, 5)
insert(root, 15)
insert(root, 30)
insert(root, 25)
insert(root, 35)

a = 5
b = 55
print(findDistWrapper(root, 5, 35))
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.prev = None
        self.next = None

def convertToDoublyLinkedList(root):
    if root is None:
        return None


    head = None
    tail = None

    def inorderTraversal(node):
        nonlocal head, tail

        if node is None:
            return

        inorderTraversal(node.left)

        if head is None:
            head = node
        else:
            tail.next = node
            node.prev = tail

        tail = node

        inorderTraversal(node.right)


    inorderTraversal(root)

    return head



tree = TreeNode(10)
tree.left = TreeNode(5)
tree.right = TreeNode(20)
tree.right.left = TreeNode(30)
tree.right.right = TreeNode(35)


doublyLinkedListHead = convertToDoublyLinkedList(tree)


def printDoublyLinkedList(head):
    if head is None:
        return

    current = head

    while current is not None:
        print(current.value, end=' ')
        current = current.next

printDoublyLinkedList(doublyLinkedListHead)  # Output: 5 10 30 20 35
```

```python
class Node:

    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        self.next = None

def connectAtSameLevel(root):
    q = []
    q.append(root)

    while (len(q) > 0):
        prev, cur = None, None
        size = len(q)

        for i in range(size):

            cur = q[0]
            q.pop(0)

            if (i > 0):
                prev.next = cur

            if (cur.left != None):
                q.append(cur.left)

            if (cur.right != None):
                q.append(cur.right)

            prev = cur
        prev.next = None

def printLevelWise(root):
    cur = root
    while(cur != None):
        print(cur.val, " --> ", end = " ")
        cur = cur.next

    print("-1")

    cur = root.left
    while(cur!=None):
        print(cur.val, " --> ", end = " ")
        cur = cur.next

    print("-1")

    cur = root.left.left
    while(cur!=None):
        print(cur.val, " --> ", end = " ")
        cur = cur.next

    print("-1")



root = Node(1)
root.left =  Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

connectAtSameLevel(root)
printLevelWise(root)
```
